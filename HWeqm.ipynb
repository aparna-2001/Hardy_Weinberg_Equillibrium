```python

from matplotlib import pyplot as plt
import numpy as np



Total_population = int( input('Enter the number of total population: '))
num_AA = int(input('Enter the number of homozygous dominant: '))
num_aa = int(input('Enter the number of homozygous recessive: '))
num_Aa = int(input('Enter the number of heterozygous: '))

freq_AA = num_AA/Total_population
freq_aa = num_aa/Total_population
freq_Aa = num_Aa/Total_population

p = freq_AA + 0.5*freq_Aa
q = freq_aa + 0.5*freq_Aa


homo_dom = p**2
homo_rec = q**2
hetero = 2 * p * q
ans = homo_dom + hetero + homo_rec
print(ans)

if ans == 1:
   print('The population is in Hardy Weinberg Equilibrium')
else:
   print('The population is not in Hardy Weinberg Equilibrium')


#observed and expected frequencies

observed_AA = num_AA
expected_AA = homo_dom * Total_population

observed_aa = num_aa
expected_aa = homo_rec * Total_population

observed_Aa = num_Aa
expected_Aa = hetero * Total_population

print(f"observed genotype frequency of homozygous dominant is {observed_AA} and expected genotype frequency is {expected_AA}" )

print(f"observed genotype frequency of homozygous recessive is {observed_aa} and expected genotype frequency is {expected_aa}" )

print(f"observed genotype frequency of heterozygous is {observed_Aa} and expected genotype frequency is {expected_Aa}" )



#visualisation

def graph(p, homo_dom, hetero, homo_rec):
    # theoretical curve
    p_range = np.linspace(0,1,100)
    q_range = 1 - p_range
    plt.plot(p_range, p_range**2, label='AA Expected Curve', linewidth=2, color='blue')
    plt.plot(p_range, 2*p_range*q_range, label='Aa Expected Curve', linewidth=2, color='green')
    plt.plot(p_range, q_range**2, label='aa Expected Curve', linewidth=2, color='yellow')

    # user data
    plt.scatter(p, homo_dom, color='red', s=70, label='User AA Value')
    plt.scatter(p, hetero, color='red', s=70, label='User Aa Value')
    plt.scatter(p, homo_rec, color='red', s=70, label='User aa Value')

    plt.xlabel('Allele frequency (p)')
    plt.ylabel('Genotype frequency')
    plt.legend()
    plt.show()

graph(p, homo_dom, hetero, homo_rec)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Hardy-Weinberg Equilibrium equations:
# p + q = 1 (allele frequencies)
# p + 2pq + q = 1 (genotype frequencies)
# where p = freq of dominant allele, q = freq of recessive allele

# Create range of allele frequencies for p (from 0 to 1)
p = np.linspace(0, 1, 100)
q = 1 - p

# Calculate genotype frequencies
freq_AA = p**2      # Homozygous dominant
freq_Aa = 2*p*q     # Heterozygous
freq_aa = q**2      # Homozygous recessive

# Create the plot
plt.figure(figsize=(10, 6))

plt.plot(p, freq_AA, label='AA (p)', linewidth=2, color='blue')
plt.plot(p, freq_Aa, label='Aa (2pq)', linewidth=2, color='green')
plt.plot(p, freq_aa, label='aa (q)', linewidth=2, color='red')

plt.xlabel('Frequency of A allele (p)', fontsize=12)
plt.ylabel('Genotype Frequency', fontsize=12)
plt.title('Hardy-Weinberg Equilibrium', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.xlim(0, 1)
plt.ylim(0, 1)

# Add annotation for a specific example
example_p = 0.7
example_q = 0.3
plt.axvline(x=example_p, color='gray', linestyle='--', alpha=0.5)
plt.text(example_p + 0.02, 0.9, f'Example:\np = {example_p}\nAA = {example_p**2:.3f}\nAa = {2*example_p*example_q:.3f}\naa = {example_q**2:.3f}',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

# Optional: Create a stacked area chart
plt.figure(figsize=(10, 6))

plt.fill_between(p, 0, freq_aa, label='aa (q)', alpha=0.7, color='red')
plt.fill_between(p, freq_aa, freq_aa + freq_Aa, label='Aa (2pq)', alpha=0.7, color='green')
plt.fill_between(p, freq_aa + freq_Aa, 1, label='AA (p)', alpha=0.7, color='blue')

plt.xlabel('Frequency of A allele (p)', fontsize=12)
plt.ylabel('Cumulative Genotype Frequency', fontsize=12)
plt.title('Hardy-Weinberg Equilibrium - Stacked View', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='center left')
plt.grid(True, alpha=0.3, axis='x')
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```
