```python



Total_population = int( input('Enter the number of total population: '))
num_AA = int(input('Enter the number of homozygous dominant: '))
num_aa = int(input('Enter the number of homozygous recessive: '))
num_Aa = int(input('Enter the number of heterozygous: '))

freq_AA = num_AA/Total_population
freq_aa = num_aa/Total_population
freq_Aa = num_Aa/Total_population

p = freq_AA + 0.5*freq_Aa
q = freq_aa + 0.5*freq_Aa


homo_dom = p**2
homo_rec = q**2
hetero = 2 * p * q
ans = homo_dom + hetero + homo_rec
print(ans)

if ans == 1:
   print('The population is in Hardy Weinberg Equilibrium')
else:
   print('The population is not in Hardy Weinberg Equilibrium')


#observed and expected frequencies

observed_AA = num_AA
expected_AA = homo_dom * Total_population

observed_aa = num_aa
expected_aa = homo_rec * Total_population

observed_Aa = num_Aa
expected_Aa = hetero * Total_population

print(f"observed genotype frequency of homozygous dominant is {observed_AA} and expected genotype frequency is {expected_AA}" )

print(f"observed genotype frequency of homozygous recessive is {observed_aa} and expected genotype frequency is {expected_aa}" )

print(f"observed genotype frequency of heterozygous is {observed_Aa} and expected genotype frequency is {expected_Aa}" )
```

```python
#allelic frequency and genotype frequency visualization

from matplotlib import pyplot as plt


allele_freq = [0.2, 0.8]
genotype_freq = [0.04, 0.64]



def graph(allele_freq,genotype_freq):
    plt.xlabel('Allele frequency')
    plt.ylabel('Genotype frequency')
    plt.plot(allele_freq,genotype_freq)
    plt.show()

graph(allele_freq,genotype_freq)
```
